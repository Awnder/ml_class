```python
from matplotlib import pyplot as plt
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
df_train = pd.read_csv('housing_data.csv')
df_test = pd.read_csv('housing_test.csv')
# neighborhood data
```

```python
df_train.head()
# df_test.head()
```

```python
df_train.dropna(inplace=True)
df_test.dropna(inplace=True)
y_train = df_train[['median_house_value']]
X_train = df_train.drop(columns=['median_house_value'])
X_test = df_test
```

```python
# init standardizations
scalar = MinMaxScaler()
scalar.fit(X_train[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households']])
```

```python
encoded_data = scalar.transform(X_train[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households']])
scalared_df = pd.DataFrame(encoded_data, columns=['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households'])
scalared_df.head()
```

```python
hotencoder = OneHotEncoder(handle_unknown='ignore')
encoded_data = hotencoder.fit_transform(X_train[['ocean_proximity']])
hot_data_df = pd.DataFrame(encoded_data.toarray(), columns=hotencoder.get_feature_names_out()).astype('int8')
hot_data_df.head()
```

```python
X_train_normalized = pd.concat([scalared_df, hot_data_df], axis=1)
X_train_normalized.head()
```

```python
encoded_data = hotencoder.transform(X_test[['ocean_proximity']])
hot_data_df = pd.DataFrame(encoded_data.toarray(), columns=hotencoder.get_feature_names_out()).astype('int8')
hot_data_df.head()
```

```python
encoded_data = scalar.transform(X_test[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households']])
scalared_df = pd.DataFrame(encoded_data, columns=['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households'])
scalared_df.head()
```

```python
X_test_normalized = pd.concat([scalared_df, hot_data_df], axis=1)
X_test_normalized.head()
```

```python
regression = LinearRegression().fit(X_train_normalized, y_train)
```

```python
prediction = regression.predict(X_test_normalized)
```
