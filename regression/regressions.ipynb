```python
from matplotlib import pyplot as plt
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
# With Aid from Perplexity

filename = 'regression1.csv'

# Extract x and y values from the CSV data
x = []
y = []

# Skip the header row
data = open(filename).read()    

data = data[4:-1] # removing the x,y and empty cell at end
for line in data.split('\n'):
    values = line.split(',')
    x.append(float(values[0]))
    y.append(float(values[1]))
```

```python

# Convert lists to numpy arrays
X = np.array(x).reshape(-1, 1)  # reshape to 2D array for sklearn
y = np.array(y)

# Now X and y can be used for sklearn's linear regression
model = LinearRegression()
model.fit(X, y)
m, b = model.coef_[0], model.intercept_
print(m,b)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.scatter(X, y, color='blue')
ax.set_title('Regression 1')
ax.plot(X, m*X + b, color='red')
```

```python
# Code from Perplexity
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

filename = 'regression2.csv'

# Extract x and y values from the CSV data
x = []
y = []

# Skip the header row
data = open(filename).read()    

data = data[4:-1] # removing the x,y and empty cell at end
for line in data.split('\n'):
    values = line.split(',')
    x.append(float(values[0]))
    y.append(float(values[1]))

# Convert lists to numpy arrays
X = np.array(x).reshape(-1, 1)  # reshape to 2D array for sklearn
y = np.array(y)

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Now X and y can be used for sklearn's linear regression
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X, y)
# m, b = model.coef_[0], model.intercept_
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Polynomial Regression')

# Add simple linear regression for comparison
linear_model = LinearRegression().fit(X, y)
y_linear_pred = linear_model.predict(X)
plt.plot(X, y_linear_pred, color='green')

plt.title('Polynomial Regression')
plt.show()
```
